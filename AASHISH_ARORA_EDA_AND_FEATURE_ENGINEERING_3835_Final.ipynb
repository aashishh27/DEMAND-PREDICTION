```python
import pandas as pd
import numpy as np
import re
import json

def extract_fsa_from_address(row):
    """
    For each row in the Clients DataFrame, this function attempts to extract
    the FSA (Forward Sortation Area - the first three characters of the postal code)
    from available address fields in the following order:
      1. 'address_text'
      2. 'address'
      3. 'zz_address_txt' (JSON formatted)
    It uses a regex pattern to capture an Edmonton postal code (e.g., "T5A 3X2") and returns
    the first three characters (e.g., "T5A"). If nothing is found, returns "Unknown".
    """
    # Regex pattern: captures the FSA part from postal codes like "T5A 3X2" or "T6B3L7"
    pattern = r'\b(T[56][A-Z])\s?\d[A-Z]\d\b'

    # Priority 1: 'address_text'
    addr_text = row.get('address_text')
    if pd.notnull(addr_text) and isinstance(addr_text, str) and addr_text.strip():
        match = re.search(pattern, addr_text, re.IGNORECASE)
        if match:
            return match.group(1).upper()

    # Priority 2: 'address'
    addr = row.get('address')
    if pd.notnull(addr) and isinstance(addr, str) and addr.strip():
        match = re.search(pattern, addr, re.IGNORECASE)
        if match:
            return match.group(1).upper()

    # Priority 3: 'zz_address_txt' (assumed JSON formatted)
    zz_addr = row.get('zz_address_txt')
    if pd.notnull(zz_addr) and isinstance(zz_addr, str) and zz_addr.strip():
        try:
            # Clean the JSON string
            cleaned = zz_addr.strip().replace("\n", " ").replace("\r", " ")
            data = json.loads(cleaned)
            pc = data.get('postal_code', "")
            if pc:
                match = re.search(pattern, pc, re.IGNORECASE)
                if match:
                    return match.group(1).upper()
        except Exception:
            # Fall back to raw text matching if JSON parsing fails.
            match = re.search(pattern, zz_addr, re.IGNORECASE)
            if match:
                return match.group(1).upper()
    return "Unknown"

# Apply the function row-wise to produce a new FSA column in clients_df_clean.
clients_df_clean['FSA'] = clients_df_clean.apply(extract_fsa_from_address, axis=1)

# Because some unique clients might have multiple rows, group by "unique id" and choose the first non-"Unknown" FSA.
def pick_first_fsa(series):
    valid = series[series != "Unknown"]
    if not valid.empty:
        return valid.iloc[0]
    else:
        return "Unknown"

fsa_mapping = clients_df_clean.groupby('unique id')['FSA'] \
                              .apply(pick_first_fsa) \
                              .reset_index() \
                              .rename(columns={'FSA': 'final_FSA'})

# Merge the FSA mapping back into clients_df_clean so that each unique client has one final FSA.
clients_df_clean = clients_df_clean.merge(fsa_mapping, on='unique id', how='left')

# Verify by printing a sample mapping of unique id and final_FSA.
print("Sample FSAs for unique clients:")
print(fsa_mapping.head(10))

# Count how many unique clients still have "Unknown" as their final FSA.
unknown_count = fsa_mapping[fsa_mapping['final_FSA'] == "Unknown"]['unique id'].nunique()
total_unique = fsa_mapping['unique id'].nunique()
print("\nNumber of unique clients with 'Unknown' final FSA:", unknown_count)
print("Percentage: {:.2f}%".format((unknown_count / total_unique) * 100))
```

```python
import numpy as np
import pandas as pd

# Define the two IFSSA centre FSAs.
ifssa_centre_fsa = ["T5H", "T6E"]

# Function to assign a random FSA from the known centres if the extracted FSA is "Unknown"
def assign_random_ifssa_fsa(fsa):
    if pd.isnull(fsa) or fsa.strip().lower() == "unknown":
        return np.random.choice(ifssa_centre_fsa)
    return fsa

# Assume your clients_df_clean already has a column 'final_FSA' with the extracted FSA from the combined address fields.
# Apply the function to fill "Unknown" values with a random known IFSSA centre FSA.
clients_df_clean['final_FSA'] = clients_df_clean['final_FSA'].apply(assign_random_ifssa_fsa)

# Verify by printing a sample of the final FSA for unique clients.
fsa_mapping = clients_df_clean[['unique id', 'final_FSA']].drop_duplicates()
print("Sample FSAs for unique clients after assigning random IFSSA centre FSA for 'Unknown':")
print(fsa_mapping.head(10))

# Count how many unique clients still have "Unknown" as the final FSA (should be 0 after assignment)
unknown_count = fsa_mapping[fsa_mapping['final_FSA'].str.strip().str.lower() == "unknown"]['unique id'].nunique()
total_unique = fsa_mapping['unique id'].nunique()
print("\nNumber of unique clients with 'Unknown' final FSA after assignment:", unknown_count)
print("Percentage: {:.2f}%".format((unknown_count / total_unique) * 100))

```

We began by extracting the Forward Sortation Area (FSA), which comprises the first three characters of a Canadian postal code, from our client address data. Our approach prioritized the "address_text" field first, then the "address" field, and lastly attempted to extract the postal code from the JSON-formatted "zz_address_txt" field using a regular expression. For each client, we grouped the data by their unique identifier and selected the first non-"Unknown" FSA value to represent that client’s final FSA. Recognizing that some clients still ended up with an "Unknown" FSA—likely due to incomplete or inconsistent address information—we then replaced those unknown values by randomly assigning one of two known IFSSA centre FSAs (either "T5H" or "T6E") as a fallback. Finally, we verified the results by creating a mapping of unique client IDs to their final FSAs and then counted the number of unique FSAs present in the dataset. This process ensured that every client had a valid FSA, which can later be used for spatial analysis and mapping of food hamper demand.

```python
# Step 1: Group pickups by date + FSA
daily_fsa_demand = hamper_data.groupby(['pickup_date', 'final_FSA']).size().reset_index(name='daily_pickups')

# Step 2: Sort for time series
daily_fsa_demand = daily_fsa_demand.sort_values(['final_FSA', 'pickup_date'])

# Step 3: Create rolling features
for window in [7, 14, 30]:
    daily_fsa_demand[f'rolling_{window}d'] = (
        daily_fsa_demand
        .groupby('final_FSA')['daily_pickups']
        .transform(lambda x: x.shift(1).rolling(window).sum())
    )

# Optional: Merge rolling data back into hamper_data
hamper_data = hamper_data.merge(
    daily_fsa_demand,
    how='left',
    on=['pickup_date', 'final_FSA']
)
```

```python
def get_season(month):
    if month in [12, 1, 2]: return 'Winter'
    elif month in [3, 4, 5]: return 'Spring'
    elif month in [6, 7, 8]: return 'Summer'
    else: return 'Fall'

hamper_data['season'] = hamper_data['pickup_month'].apply(get_season)
```

##  Block 3: Client & Household Feature Engineering

To support regional demand prediction, we engineered client behavior features that can be **aggregated by region** to reveal growth, stability, or change in service usage.

---

###  Key Features Created

| Feature | Description | Relevance |
|--------|-------------|-----------|
| `first_visit_date` | Date of each client’s first pickup | Used to calculate client engagement age |
| `days_since_first_visit` | Number of days since first visit at time of each pickup | Tracks client maturity |
| `new_client_flag` | 1 if client had joined within the past 30 days | Helps detect recent outreach impact |
| `visit_count_90d` | Number of visits made by the client in the past 90 days | Captures active usage intensity |
| `region_new_client_ratio_30d` | Rolling 30-day ratio of visits from new clients by region | Signals areas with growing new demand |

---

###  Visual Insights

- **Boxplot of 90-day rolling visits** helps identify **regions with higher engagement intensity**.
- **New client ratio over time** shows which areas are gaining or retaining **new families**, which helps prioritize **outreach and delivery planning**.

---

###  Summary

These features allow us to:
- Capture real-time **client behavior changes per region**
- Spot regions with **rising demand** due to new enrollments
- Use **client-level data to power region-level predictions**

All engineered variables contribute directly to the **geospatial demand modeling goal**.


```python
# Copy base DataFrame
region_client_df = hamper_data.copy()
region_client_df['pickup_date'] = pd.to_datetime(region_client_df['pickup_date'], errors='coerce')

# === 1. First visit date per client ===
first_visits = region_client_df.groupby('unique_client')['pickup_date'].min().reset_index()
first_visits.rename(columns={'pickup_date': 'first_visit_date'}, inplace=True)
region_client_df = region_client_df.merge(first_visits, on='unique_client', how='left')

# === 2. Days since first visit ===
region_client_df['days_since_first_visit'] = (
    region_client_df['pickup_date'] - region_client_df['first_visit_date']
).dt.days

# === 3. New client flag (first 30 days of activity) ===
region_client_df['new_client_flag'] = (region_client_df['days_since_first_visit'] <= 30).astype(int)

# === 4. Visit count in past 90 days (per client) ===
region_client_df = region_client_df.sort_values(by=['unique_client', 'pickup_date'])
rolling_counts = []

for client_id, group in region_client_df.groupby('unique_client'):
    group = group.sort_values('pickup_date')
    count_list = []
    for i in range(len(group)):
        current_date = group.iloc[i]['pickup_date']
        past_window = group[
            (group['pickup_date'] >= current_date - pd.Timedelta(days=90)) &
            (group['pickup_date'] <= current_date)
        ]
        count_list.append(len(past_window))
    rolling_counts.extend(count_list)

region_client_df['visit_count_90d'] = rolling_counts

# === 5. Region-level % of new clients (30-day rolling window) ===
daily_region = (
    region_client_df
    .groupby(['pickup_date', 'region'])['new_client_flag']
    .agg(['sum', 'count'])
    .reset_index()
    .rename(columns={'sum': 'new_clients', 'count': 'total_visits'})
)

daily_region = daily_region.sort_values(['region', 'pickup_date'])
daily_region['region_new_client_ratio_30d'] = (
    daily_region.groupby('region')['new_clients'].transform(lambda x: x.rolling(window=30, min_periods=1).sum()) /
    daily_region.groupby('region')['total_visits'].transform(lambda x: x.rolling(window=30, min_periods=1).sum())
)

# Merge back into main DataFrame
region_client_df = region_client_df.merge(
    daily_region[['pickup_date', 'region', 'region_new_client_ratio_30d']],
    on=['pickup_date', 'region'],
    how='left'
)

# === Final Preview ===
region_client_df[['unique_client', 'pickup_date', 'days_since_first_visit',
                  'visit_count_90d', 'new_client_flag', 'region_new_client_ratio_30d']].head()
```

```python
# ─── Block 4: Geospatial Features ───────────────────────────────────────────────
import pandas as pd
import pgeocode
from geopy.distance import geodesic
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1) Lookup FSA centroids
nomi = pgeocode.Nominatim('CA')
unique_fsas = region_client_df['final_FSA'].unique().tolist()
fsa_geo = (
    nomi.query_postal_code(unique_fsas)
        [['postal_code','latitude','longitude']]
        .dropna()
        .rename(columns={'postal_code':'final_FSA'})
)

# 2) Merge lat/lon into your feature table
region_client_df = region_client_df.merge(fsa_geo, on='final_FSA', how='left')

# 3) Compute distance to Edmonton hub
hub = (53.5444, -113.4909)
region_client_df['dist_to_hub_km'] = region_client_df.apply(
    lambda r: geodesic((r.latitude, r.longitude), hub).km
              if pd.notnull(r.latitude) else None,
    axis=1
)

# 4) FSA‐level 14‑day pickup total (neighborhood demand)
fsa_demand = (
    region_client_df
    .groupby('final_FSA')['target_pickup_count_14d']
    .sum()
    .reset_index(name='fsa_total_14d_pickups')
)
region_client_df = region_client_df.merge(fsa_demand, on='final_FSA', how='left')

# 5) Cluster FSAs into 5 geo‑zones
coords = fsa_geo[['latitude','longitude']].values
km = KMeans(n_clusters=5, random_state=42).fit(coords)
fsa_geo['fsa_cluster'] = km.labels_
region_client_df = region_client_df.merge(
    fsa_geo[['final_FSA','fsa_cluster']],
    on='final_FSA', how='left'
)

# 6) Visualize the clusters
plt.figure(figsize=(6,6))
plt.scatter(fsa_geo.longitude, fsa_geo.latitude,
            c=fsa_geo.fsa_cluster, cmap='tab10', s=50)
plt.title('FSA Centroid Clusters (k=5)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
